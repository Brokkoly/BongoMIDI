```python
from mido import MidiFile
import numpy as np
mid = MidiFile('d_e1m1.mid')
```

```python
def splitMessage(msg):#command,channel,note,velocity,time):
    channel = msg[1].split("=")#channel.split("=")[1]
    note = msg[2].split("=")[1]
    velocity = msg[3].split("=")[1]
    time = msg[4].split("=")[1]
    return [msg[0],channel,note,velocity,time]
```

```python
alltracks=[[]]
for i,track in enumerate(mid.tracks):
    print('Track {}: {}'.format(i,track.name))
    alltracks.append([])
    for msg in track:
        alltracks[i].append(msg)
        print(msg)
```

```python
track2 = np.empty((0))
msg=""
for msgNum in range(np.size(alltracks[2])):
    msg = str(alltracks[2][msgNum]).split(" ")
    if(msg[0]=="meta"):
        continue
    if(msg[0]=="program_change"):
        continue
    print(msg)
    msg=splitMessage(msg)
    print(msg)
    track2 = np.append(track2,msg)
    
track2 = np.reshape(track2,(np.size(track2)//4,4))
print(track2[1])
```
